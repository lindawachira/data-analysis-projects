```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
num_of_pages = 2
rows_per_page = 50
urls =[f"https://myanimelist.net/topanime.php?type=movie&limit={i * rows_per_page}" for i in range(num_of_pages)]
```

```python
pages = []
```

```python
for url in urls:
  print(f"Scraping: {url}")
  page = requests.get(url)
  soup = BeautifulSoup(page.content, "html.parser")
  pages.append(soup)
  print(pages)
```

```python
rows = []
for soup in pages:
    table = soup.find("table", class_="top-ranking-table")
    for row in table.find_all("tr", class_="ranking-list"):

      rowTitle1 = row.find("td",class_= "title al va-t word-break")
      rowTitle2= rowTitle1.find("div", class_="detail")
      rowTitle3 = rowTitle2.find("div", class_="di-ib clearfix")
      RowTitle = rowTitle3.find("h3").text.strip()

      RowRank = row.find("td",class_="rank ac").text.strip()
      RowScore = row.find("td",class_="score ac fs14").text.strip()

     # Get Image URL
      img_tag = row.find("img")
      image_url = img_tag.get("data-src") if img_tag else None

      print(RowTitle, RowRank, RowScore, image_url)
      row_data = [RowTitle, RowRank, RowScore, image_url]
      rows.append(row_data)
print(rows)
```

```python
columns = ["Title", "Rank", "Score", "Image"]
```

```python
df = pd.DataFrame(rows, columns=columns)
df.to_csv(r"Top100_Anime_Movies.csv",index=False,encoding="utf-8")
```
